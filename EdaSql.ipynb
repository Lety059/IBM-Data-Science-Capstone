```python
import sqlite3
import csv
import pandas as pd

# Crea una connessione a SQLite
conn = sqlite3.connect('spacex.db')
cursor = conn.cursor()
```

```python
# Creazione della tabella SpaceX in SQLite
create_ = '''
            DROP TABLE IF EXISTS SpaceX;
            CREATE TABLE SpaceX
                (
                    Date DATE NULL,
                    Time TIME NULL,
                    BoosterVersion VARCHAR(50) NULL,
                    LaunchSite VARCHAR(50) NULL,
                    Payload VARCHAR(100) NULL,
                    PayloadMassKG INT NULL,
                    Orbit VARCHAR(50) NULL,
                    Customer VARCHAR(100) NULL,
                    MissionOutcome VARCHAR(50) NULL,
                    LandingOutcome VARCHAR(100) NULL
                );
            '''
cursor.execute(create_)
conn.commit()
print('La tabella SpaceX è stata creata con successo!')
```

```python
# Creazione della tabella SpaceX in SQLite
create_table = '''
            DROP TABLE IF EXISTS SpaceX;
            CREATE TABLE SpaceX
                (
                    Date DATE NULL,
                    Time TIME NULL,
                    BoosterVersion VARCHAR(50) NULL,
                    LaunchSite VARCHAR(50) NULL,
                    Payload VARCHAR(100) NULL,
                    PayloadMassKG INT NULL,
                    Orbit VARCHAR(50) NULL,
                    Customer VARCHAR(100) NULL,
                    MissionOutcome VARCHAR(50) NULL,
                    LandingOutcome VARCHAR(100) NULL
                );
            '''

# Split the SQL statements and execute them one by one
for statement in create_table.split(';'):
    if statement.strip():  # Execute only if the statement is not empty
        cursor.execute(statement)
conn.commit()
print('La tabella SpaceX è stata creata con successo!')
```

```python
query_3 = '''
    SELECT SUM(PAYLOAD_MASS__KG_) AS Total_PayloadMass
    FROM spacex_data
    WHERE Customer LIKE 'NASA (CRS)'
'''
result_3 = pd.read_sql(query_3, conn)
print(result_3)
```

```python
query_3 = '''
    SELECT SUM(PAYLOAD_MASS__KG_) AS Total_PayloadMass
    FROM spacex_data
    WHERE Customer LIKE 'NASA (CRS)'
'''
result_3 = pd.read_sql(query_3, conn)
print(result_3)
```

```python
query_4 = '''
    SELECT AVG(PAYLOAD_MASS__KG_) AS Avg_PayloadMass
    FROM spacex_data
    WHERE Booster_Version = 'F9 v1.1'
'''
result_4 = pd.read_sql(query_4, conn)
print(result_4)
```

```python
query_5 = '''
    SELECT MIN(Date) AS FirstSuccessfull_landing_date
    FROM spacex_data
    WHERE Landing_Outcome LIKE 'Success (ground pad)'
'''
result_5 = pd.read_sql(query_5, conn)
print(result_5)
```

```python
query_6 = '''
    SELECT Booster_Version
    FROM spacex_data
    WHERE Landing_Outcome = 'Success (drone ship)'
      AND PAYLOAD_MASS__KG_ > 4000
      AND PAYLOAD_MASS__KG_ < 6000
'''
result_6 = pd.read_sql(query_6, conn)
print(result_6)
```

```python
query_7a = '''
    SELECT COUNT(Mission_Outcome) AS SuccessOutcome
    FROM spacex_data
    WHERE Mission_Outcome LIKE 'Success%'
'''
result_7a = pd.read_sql(query_7a, conn)
print("Total Successes:", result_7a)
```

```python
query_7b = '''
    SELECT COUNT(Mission_Outcome) AS FailureOutcome
    FROM spacex_data
    WHERE Mission_Outcome LIKE 'Failure%'
'''
result_7b = pd.read_sql(query_7b, conn)
print("Total Failures:", result_7b)
```

```python
query_8 = '''
    SELECT Booster_Version, PAYLOAD_MASS__KG_
    FROM spacex_data
    WHERE PAYLOAD_MASS__KG_ = (
                            SELECT MAX(PAYLOAD_MASS__KG_)
                            FROM spacex_data
                          )
    ORDER BY Booster_Version
'''
result_8 = pd.read_sql(query_8, conn)
print(result_8)
```

```python
query_9 = '''
    SELECT Booster_Version, Launch_Site, Landing_Outcome
    FROM spacex_data
    WHERE Landing_Outcome LIKE 'Failure (drone ship)'
      AND Date BETWEEN '2015-01-01' AND '2015-12-31'
'''
result_9 = pd.read_sql(query_9, conn)
print(result_9)
```

```python
query_10 = '''
    SELECT Landing_Outcome, COUNT(Landing_Outcome) AS OutcomeCount
    FROM spacex_data
    WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
    GROUP BY Landing_Outcome
    ORDER BY OutcomeCount DESC
'''
result_10 = pd.read_sql(query_10, conn)
print(result_10)
```
